# EDA operation of fifa

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
plt.style.use('ggplot') 
```

### Importing Dataset

```python
fifa1 = pd.read_csv("D:\\DATASCIENCE\\MyProject\\PythonProject\\DATASET\\fifa_eda.csv")
```

```python
fifa1.head(10) #shows the starting 10 from the dataset
```

```python
fifa1.describe()
```

```python
#To analyse missing values and datatypes from the datasets
fifa1_missing = fifa1.isna().mean()
fifa1_dtype = fifa1.dtypes

details_1 = pd.DataFrame(fifa1_missing, columns = ["missing"])
details_2 = pd.DataFrame(fifa1_dtype, columns = ["datatype"])

details = pd.concat([details_1, details_2], axis=1)
details.sort_values(by = "missing", ascending = False)
```

```python
fifa1.info()
```

### Convert data type of column to float

```python
fifa1["Potential"] = fifa1["Potential"].astype(float)
```

```python
fifa1.info()
```

```python
fifa1.isnull().sum() #This gives the total sum of null value in each column
```

```python
#After calculating all the null value tuples , now we remove all those null tuples
fifa2 = fifa1.dropna(subset = ["Club",'Value','International Reputation','Skill Moves','Contract Valid Until'])
```

```python
fifa2.isnull().sum() 
```

```python
#Check whether duplicate data exists
fifa1.duplicated().any()
```

### Correlation

 heatmap to check how closely related our categorical values are to each other

```python
fifa1_corr = fifa2[['Overall','Value','Wage']].corr()
fifa1_corr
```

```python
sns.heatmap(fifa1_corr,annot=True) #Heat map
```

Value" is pretty closely related to "Wage", so lets visualize this using scatterplot.

```python
ab = sns.scatterplot(x='Wage',
               y="Value",
              data=fifa2)
ab.set_title("Value vs Wage")
plt.show()
```
